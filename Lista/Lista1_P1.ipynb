# Função para normalizar a série entre 0 e 1

## A função usada para normalizar a série é:
## $A_{norm} = \frac{A_i - min(A)}{max(A) - min(A)}$

```python
def Normalizar(data):
    serie_norm = np.zeros([len(data),10])
    Promedio = np.zeros([10])
    Varianza = np.zeros([10])
    Skewness = np.zeros([10])
    Kurtosis = np.zeros([10])
    
    for t in range(10):
        for i in range(len(data)):
            serie_norm[i,t] = (data[i,t] - np.min(data[:,t]))/(np.max(data[:,t]) - np.min(data[:,t]))
    
        Promedio[t] = np.mean(serie_norm[:,t])
        Varianza[t] = np.var(serie_norm[:,t])
        Skewness[t] = skew(serie_norm[:,t])
        Kurtosis[t] = kurtosis(serie_norm[:,t])
        
    if len(data) == 64:
        color = 'bo'
    elif len(data) == 128:
        color = 'ro'
    elif len(data) == 256:
        color = 'go'
    elif len(data) == 512:
        color = 'ko'
    elif len(data) == 1024:
        color = 'co'
    elif len(data) == 2048:
        color = 'mo'
    elif len(data) == 4096:
        color = 'yo'
    elif len(data) == 8192:
        color = 'b*'
        
    serie_esta = pd.DataFrame({'Group':str(len(data)),'Prom':Promedio,'Vari':Varianza, \
                                'Asim':Skewness,'Curt':Kurtosis,'Cor':color})    
        
    return serie_norm,serie_esta
```

```python
serie_64_norm,serie_64_esta = Normalizar(serie_64)
serie_128_norm,serie_128_esta = Normalizar(serie_128)
serie_256_norm,serie_256_esta = Normalizar(serie_256)
serie_512_norm,serie_512_esta = Normalizar(serie_512)
serie_1024_norm,serie_1024_esta = Normalizar(serie_1024)
serie_2048_norm,serie_2048_esta = Normalizar(serie_2048)
serie_4096_norm,serie_4096_esta = Normalizar(serie_4096)
serie_8192_norm,serie_8192_esta = Normalizar(serie_8192)
```

# K-means

### Para aplicar o método K-means aos dados, primeiro é necessário conhecer o número adequado de grupos, para isso é utilizado a grafica Elbow Curve.

```python
X = np.array(result[["Vari","Asim","Curt"]])
y = np.array(result['Group'])
X.shape
```

```python
Nc = range(1,16)
kmeans = [KMeans(n_clusters=i) for i in Nc]
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
```

```python
fig3 = plt.figure(figsize=(8,6),dpi=100)
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.grid(True)
plt.show()
```
